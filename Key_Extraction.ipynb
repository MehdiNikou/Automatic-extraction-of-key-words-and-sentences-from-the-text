## Rank k approximation

```python
K = 5 
NMF = NMF(n_components=K, init='random', random_state=0)
C = NMF.fit_transform(term_sentance_matrix)
D = NMF.components_
```

```python
C.shape
```

```python
D.shape
```

### function that calculate norm 2 of columns of a matrix and return number of column with maximum norm

```python
def arg_max(D,roundd):
    list=[]
    for i in range(D.shape[1]):
        list.append(np.linalg.norm(D[roundd:,i]))
    return np.argmax(list)
```

### a key sentence with rank k approximation

```python
arg_max_D=arg_max(D,0)
arg_max_D
```

```python
text[arg_max_D]
```

```python
def swap(D,top_sentence,roundd):
    I=np.eye(D.shape[1])
    I[:, [roundd, top_sentence]] = I[:, [top_sentence, roundd]]
    P=I
    DP=np.dot(D, P)
    return DP
```

```python
DP1 =swap(D,arg_max_D,0)
```

### function for sign

```python
def sign(x):
    if x>=0:
        return 1
    else:
        return -1
```

### function for house holder matrix

```python
def house_holder(DP, roundd, k):
    X = np.reshape(DP[roundd:,roundd], (DP.shape[0]-roundd, 1))
    e = np.eye(K-roundd, 1)
    V = X+(sign(X[0])*(np.linalg.norm(X)*e))
    I = np.eye(K-roundd)
    H= I - 2*(np.dot(V,V.T)) / np.dot(V.T,V)
    Q = np.eye(K)
    Q[roundd:,roundd:] = H
    return Q
```

```python
Q1=house_holder(DP1,0, K)
```

```python
DP1.shape
```

### function for calculate new C and D

```python
def CD_new(C,Q,DP,roundd, K):
    Q_DP=np.dot(Q.T,DP)
    C_Q=np.dot(C,Q)
    I=np.eye(K)
    I[roundd, roundd]=Q_DP[roundd, roundd]
    T=I
    C_new=np.dot(C_Q,T)
    D_new=np.dot(np.linalg.inv(T),Q_DP)
    return C_new, D_new
```

```python
C1,D1=CD_new(C,Q1,DP1,0, K)
```

```python
C1.shape
```

```python
from sklearn.decomposition import NMF
K = 5
NMF = NMF(n_components=K, init='random', random_state=0)
C = NMF.fit_transform(term_sentance_matrix)
D = NMF.components_

list1 = []
for i in range(5):
    top_sentence_index = arg_max(D, i)
    list1.append(top_sentence_index)
    DP = swap(D, top_sentence_index, i)
    Q = house_holder(DP, i, K)
    C ,D = CD_new(C, Q, DP, i, K)
```
